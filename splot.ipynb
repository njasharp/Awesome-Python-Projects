```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student.csv')
df.head()
```

```python
sns.histplot(data=df, x='math_score')
plt.show()
```

```python
sns.histplot(data=df, x='math_score', kde=True)
plt.show()
```

```python
sns.histplot(data=df, x='math_score', kde=True , hue='gender',
             palette=['lightpink', 'blue'])
plt.show
```
